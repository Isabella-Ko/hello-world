# read from a text file as a list

```python
with open('new_f.txt','r') as f1:  # generate the handler
    lines = f1.read().splitlines()  # list : elements without '\n'

with open('new_f1.txt','w') as new_f1:
    for i in range(len(lines)): 
        new_f1.write(lines[i])
        new_f1.write('\n')
```

# Text Parsing

```python
with open('new_f.txt','r') as f_tp:
    lines2 = f_tp.readlines()

new_l = []
for i in range(len(lines2)):
    s = lines2[i].strip()
    e = s.replace(';','.')
    new_l.append(e)

index = 0
while index < len(new_l):
    if new_l[index] == '':
        del new_l[index]
        index -= 1
    index += 1
print(new_l)

with open('text_p','w') as f_tpw:
    for i in range(len(new_l)):
        f_tpw.write(new_l[i])
        f_tpw.write('\n')
```

# Quiz Game

```python
import random
def f_read(fname):
    with open(fname,'r') as f:
        content = f.readlines()
    
    return content 

countries = f_read('listOfCountries.txt')
capitals = f_read('listOfCapitals.txt')

c = {}

for i in range(len(countries)):
    c[countries[i].rstrip()] = capitals[i].rstrip()

print(c.items())
```

```python
# create the quiz file: 
num = 0
q_num = 0
for num in range(5):
    with open('capital_quiz%s.txt'%(num+1),'w') as q_f:
        q_f.write('Name:\n\n')
        q_f.write(' '*10 + 'Capital Quiz (Form%s)'%(num+1))
        q_f.write('\n\n')
    
        countries2 = list(c.keys())
        random.shuffle(countries2)

        for q_num in range(10):
            c_a = c[countries2[q_num]]  #berlin = c[Germany]
            wrong_a = list(c.values())
            wrong_a.remove(c_a)
            wrong_options = random.sample(wrong_a,3)
            options = wrong_options + [c_a]
            random.shuffle(options)
            
            q_f.write('%s The capital of %s is:\n'%((q_num+1),countries2[q_num]))
            for i in range(4):
                q_f.write('%s.%s'%('ABCD'[i],options[i]))
                q_f.write('\n')
            q_f.write('\n')
```
